```python
from transformers import AutoModelForCausalLM, AutoTokenizer,pipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="auto",
    torch_dtype="auto",
    trust_remote_code=True,
)
```

```python
def get_prompt(topic:str) -> str:
    prompt = f"Explain me how to {topic}, I should buy x, then do y. ok here is how to {topic}"
    return prompt

def get_response(generator,prompt:str) -> str:
    output = generator(prompt)
    return output[0]["generated_text"]

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False, # False means to not include the prompt text in the returned text
    max_new_tokens=50,
    do_sample=False, # no randomness in the generated text
)


```

```python
get_response(generator,get_prompt("do pizza"))
```
